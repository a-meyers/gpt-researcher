```python
from gpt_researcher import GPTResearcher
import markdown
import asyncio
import nest_asyncio
from gpt_researcher.utils.enum import Tone
from datetime import datetime

nest_asyncio.apply()

#==========================================================================================
async def get_report(query: str, report_type: str):
    researcher = GPTResearcher(query, 
                               report_type = "research_report", 
                               #tone = Tone.Informative,
                               report_format = "markdown",
                               verbose= False,)
    
    research_result = await researcher.conduct_research()
    report = await researcher.write_report()
    
    # Get additional information
    research_context = researcher.get_research_context()
    research_costs = researcher.get_costs()
    research_images = researcher.get_research_images()
    research_sources = researcher.get_research_sources()
    
    return report, research_context, research_costs, research_images, research_sources

#==========================================================================================
query = "What are the key futuristic technology subjects that companies should start preparing for?"
report_type = "research_report"

start_time = datetime.now()
report, context, costs, images, sources = asyncio.run(get_report(query, report_type))
end_time = datetime.now()

#==========================================================================================
# Add author and date to the report
report = f"**Author:** Anthony Meyers (anthony.meyers@aon.com)\n**Date:** {datetime.now().strftime('%B %d, %Y')}\n\n**THIS IS AI GENERATED CONTENT**\n\n" + report

# Append costs and sources to the report
total_time = end_time - start_time
report += f"\n\n**Total Time to Generate Report:** {total_time}\n"

report += f"\n\n## Research Costs\n\nThe total cost of the research was: ${costs:.2f}\n"

report += "\n## Research Sources\n"
for source in sources:
    report += f"- [{source['title']}]({source['url']})\n"

with open("my-docs/z"+query.replace("?", "")+".md", "w") as f:
    f.write(report)
```
